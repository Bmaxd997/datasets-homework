```python
labels = imdb_df['genre'].unique()
sizes = imdb_df['genre'].value_counts()


fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python

fig, ax = plt.subplots()

# Example data
people = imdb_df['director'].unique()
y_pos = np.arange(len(people))
performance = imdb_df['director'].value_counts()


ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Number of movies')
ax.set_title('How many movies in IMBD 1000 list')

plt.show()
```
